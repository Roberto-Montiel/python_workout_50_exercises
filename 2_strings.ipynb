```python
def pig_latin(word: str):
    vowels = "aeiou"

    vowels_set = {letter for letter in word if letter.lower() in vowels}

    if len(vowels_set) >= 2:
        return word + "way"
    else:
        return word[1:] + word[0] + "ay"

print(pig_latin("wind"))  
```

### **2. String Practice: Sentence Transformation**

```python
def pl_sentence(sentence: str):
    vowels = "aeiou"
    words = sentence.split(" ")
    result = []

    for word in words:
        if word[0] in vowels:
            result.append(word + "way")
        else:
            first_consonant = word[0]
            word = word[1:]
            result.append(word + first_consonant + "ay")

    return " ".join(result)

        

print(pl_sentence('this is a test translation'))
```

```python
def transpose(sentence: list):
    positions = {}
    splitting = []
    result = []

    for word in sentence:
        splitting.append(word.split())

    for word in splitting:
        for pos, s in enumerate(word):
            if pos not in positions:
                positions[pos] = [s]
            else:
                positions[pos].append(s)

    for pos in positions:
        result.append(" ".join(positions[pos]))

    return result



print(transpose(['abc def ghi', 'jkl mno pqr', 'stu vwx yz']))

# def transpose(sentence: list[str]) -> list[str]:
#     rows = [row.split() for row in sentence]
#     return [" ".join(col) for col in zip(*rows)] another more pythonic way to do it 

```

### **3. String Practice: Vowel-Based Encoding**

```python
def ubbi_dubbi(word):
    output = []

    for letter in word:
        if letter in 'aeiou':
            output.append(f'ub{letter}')
        else:
            output.append(letter)

    return ''.join(output)
    
print(ubbi_dubbi('python'))

# Using a list here is more efficient than repeated string concatenation.
# It keeps the transformation simple and easier to build step by step.
# much memory. For short strings,
# it’s not a big deal. But for long
# loops and large strings, it’s a
# bad idea. 
# Every time we compute string += something it creates a new string copying the last one so it is computational expensive 
```

### **4. Sorting a string**

```python
def strsort(word : str) -> str:
    return "".join(sorted(word))

strsort("cba")
```

```python
def phrsort(phrase: str) -> str:
    words = phrase.replace(",","").split()
    words.sort()
    return ",".join(words)

phrsort("Tom Dick Harry,")
```

```python
#Which is the last word, alphabetically, in a phrase

def last_alphabetically(phrase: str) -> str:
    words = phrase.split() #assuming no symbols
    return sorted(words)[-1]
    # or directly return max(phrase.split())
```

```python
#Which is the longest word in a phrase?

def longest(phrase: str) -> str:
    return sorted(phrase.split(), key=str.__len__)[-1]
# or return max(phrase.split(), key = len)

longest("hol que ases eeeeeee para un poco no eyretyuewtygfdu")
```
