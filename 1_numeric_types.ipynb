### **1. Number Practice: Guessing Loop**

```python
from random import randint

def guessing_game():
    number = randint(0,100)
    print(number)
    while True:
        answer = int(input("Give a guessing number from 0-100: "))
        if answer > number:
            print("Too high")
        elif answer < number:
            print("Too low")
        else:
            print("Just right")
            break
        
guessing_game()
```

### **2. Number Practice: Summation Basics**

```python

def mysum(*numbers):
    total = 0
    for number in numbers:
        total += number

    return total

mysum(*[1,2],*[3,4])


```

```python
def my_sum(numbers: list, start = 0):
    total = start
    for number in numbers:
        total += number
    
    return total 

my_sum([1,2,3,4], 9)
```

```python
# Practice goal: take a list of numeric values and return their arithmetic mean.

def mean(numbers: list):
    return sum(numbers)/len(numbers)

mean([1,2,3,4])
```

```python
def length(words : list):
    shortest_word = len(words[0])
    longest_word = len(words[0])
    letters = len(words[0])

    for i in range(1,len(words)):
        shortest_word = min(shortest_word,len(words[i]))
        longest_word = max(longest_word, len(words[i]))
        letters += len(words[i])

    return (shortest_word, longest_word, letters/len(words))


print(length(["hola", "u", "bofffff", "shhh"]))

```

```python
def objects(objects: list):
    sum = 0
    
    for object in objects:
        try:
            sum += int(object)
        except (TypeError,ValueError):
            pass

    return sum

print(objects(["1",2, "33", "a"]))
```

### **3. Input Practice: Timing Tracker**

```python

def run_timing():
    """Asks the user repeatedly for numeric input. Prints the average time an
    d number of runs."""
    total_runs = 0
    total_time = 0
    
    while True:
        one_run = input("Enter 10 km run time: ")

        if not one_run:
            break

        total_runs += 1
        total_time += float(one_run)


    print(f"Average of {total_time/total_runs:.2f}, over {total_runs} runs")

run_timing()
```

```python
def func(f: float,b: int,a: int):

    s = str(f)
    left,right = s.split(".")

    left_part = left[-b:]
    right_part = right[:a]

    return float(f"{left_part}.{right_part}")

print(func(1234.5678,2,3))
```

```python
from decimal import Decimal

def exact_decimal(s1: str,s2: str):
    n1 = Decimal(s1)
    n2 = Decimal(s2)

    return n1 + n2

print(exact_decimal("0.1","0.2"))

```

### **4. Hexadecimal output** 

```python
def hex_output():
    hex_number = input("Please type in an hexadecimal number: ")
    dec_number = 0

    for power, digit in enumerate(reversed(hex_number)):
        dec_number += int(digit,16) * (16)**power

    return dec_number

print(hex_output())
```

```python
def hex_to_decimal(hex_str):
    hex_str = hex_str.upper()
    result = 0
    position = 0
    
    
    for char in reversed(hex_str):
        if ord(char) >= 48 and ord(char) <= 57: # 0-9
            valor = ord(char) - 48
        elif ord(char) >= 65 and ord(char) <= 70: # A-F
            valor = ord(char) - 55
        else:
            raise ValueError(f"Carácter inválido: {char}")
        
        result += valor * (16 ** position)
        position += 1
        
    return result


print(hex_to_decimal("1A")) # 26
```
