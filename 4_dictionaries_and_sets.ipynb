### **1. Dictionary Practice: Simple Menu Flow**

```python
MENU = {"pizza": 9, "hamburguer":8,"kebab": 5,"soup": 10, "paella": 15, 
        "salad": 7, "chicken wings" : 6 , "pasta" : 7}

def restaurant():
    total = 0

    while True:
        order = input("Order: ")
        if order == "":
            break
        elif order in MENU:
            total += MENU[order]
            print(f"{order} costs {MENU[order]}, total is {total}")
        else:
            print(f"Sorry, we are fresh out of {order} today")

    print(f"Your total is {total}")

restaurant()
```

```python
USERNAMES = {"Juan": "egdwekdh321765321", "Nick": "dkewhg21568", "Robert": "jwhdg8764", "Patrick": "dkwewr9834", 
             "John": "qknds9846", "Jack": "pqwoi6784", "Christian": "pow2b90584"}

def logging():
    while True:
        name = input("Type in your username: ")
        if not name:
            print("Denied entry")
        elif name in USERNAMES:
            password = input("Please enter your password: ")
            if USERNAMES[name] == password:
                print("Successful loggin") 
                break
            #considering just one attempt
            else:
                print("Denied entry")
                break
        else:
            print("Please type in a valid username")

logging()
```

```python
TEMPERATURES = {
    "2026-03-20": 18,
    "2026-03-21": 15,
    "2026-03-22": 17,
    "2026-03-23": 20,
    "2026-03-24": 22,
    "2026-03-25": 19,
    "2026-03-26": 16
}

def temperatures():
    dates = list (TEMPERATURES.keys()) #date list accesible by index

    while True:
        date = input("Please type in a date: ")

        if not date:
            break
        if date in TEMPERATURES:
            print(f"Temperature at date {date} was {TEMPERATURES[date]}")

            i = dates.index(date) # entered date index

            if i > 0: # it has previous day
                prev = dates[i-1]
                print (f"Previous day ({prev}): {TEMPERATURES[prev]}")

            if i < len(dates) -1: # it has next day
                next = dates[i+1]
                print(f"Previous day ({next}): {TEMPERATURES[next]}")
        else:
            print("Please type in a valid date")

temperatures()
```

```python
from datetime import date

FAMILY = {
    "Maria": date(1975, 6, 14),
    "Carlos": date(1972, 11, 3),
    "Sofia": date(2001, 3, 28),
    "Lucia": date(2004, 9, 17),
    "Abuelo Antonio": date(1945, 2, 22),
    "Abuela Carmen": date(1948, 7, 9)
}

def days_old():
    while True:
        name = input("Please type in a member of your family: ")
        if not name:
            break
        elif name in FAMILY:
            today = date.today()
            old = abs(today - FAMILY[name]) # abs just for more security
            print(f"{name} is {old.days} days old")
        else:
            print("Please type in a valid name")
        

days_old()
```

### **2. Dictionary Practice: Rainfall Tracking**

```python
def get_rainfall():
    cities = {}

    while True:
        city = input("City: ")

        if not city:
            for c , r in cities.items():
                print(f"{c}: {r}")
            
            break
        
        rain = int(input("Rain in mm: "))

        # If the city doesn't exist it returns 0
        # If it exists, it is added to the current value
        cities[city] = cities.get(city, 0) + rain

get_rainfall()
```

```python
def get_rainfall_improved():
    cities = {}

    while True:
        city = input("City: ")

        if not city:
            for c , r in cities.items():
                print(f"{c}: \nTotal: {sum(r)}\nAverage: {sum(r)/len(r)}")
            
            break
        
        rain = int(input("Rain in mm: "))

        if city not in cities:
            cities[city] = [rain]
        else:
            cities[city].append(rain)

get_rainfall_improved()
```

### **3. Dictdiff**

```python
def dictdiff(d1: dict, d2:dict):

    solution = {}

    # assuming both dicts with same length

    if d1 == d2:
        return {}
    else:
        for v1 in d1:
            if d1.get(v1) != d2.get(v1):
                solution[v1] = [d1.get(v1),d2.get(v1)]
        
        for v2 in d2:
            if v2 not in d1:
                solution[v2] = [d1.get(v2),d2.get(v2)]

    return dict(sorted(solution.items()))

# we need to loop both dicts, a more efficient solution would be this one:
# def dictdiff(first, second):
# output = {}
# all_keys = first.keys() | second.keys()
# for key in all_keys:
# if first.get(key) != second.get(key):
# output[key] = [first.get(key),
# second.get(key)]
# return output

dictdiff( {'a':1, 'b':2, 'd':3},  {'a':1, 'b':2, 'c':4})


```

```python
def merging_dicts(*dicts: dict):
    merging = {}

    for d in dicts:
        merging.update(d)

    return merging

merging_dicts({"a": 1}, {"b": 2}, {"a": 3}, {"c": 4})

```

```python
def creating_dicts(*args):

    solution = {}
    i = 0

    if not args:
        return {}
    elif len(args) >= 2 and len(args) % 2 == 0:
        while i < len(args):
            solution[args[i]] = args[i+1]
            i += 2
        return solution
    else:
        raise ValueError("please introduce an even number of arguments")

# we could have solved it a more pythonic way instead of using the while loop like this:
# return {args[i]: args[i+1] for i in range(0, len(args), 2)}


print(creating_dicts('edad', 25, 'activo',True, 'puntos', 9.5))

```

```python
def dict_partition(d: dict, f):
    d1 = {}
    d2 = {}

    for key in d:
        if f(key, d[key]):
            d1[key] = d[key]
        else:
            d2[key] = d[key]

    return d1,d2

d = {"a": 10, "b": 3, "c": 7, "d": 1}

dict_partition(d, lambda k, v: v > 5)       
```

### **4. How many different numbers**

```python
def how_many_different_numbers(numbers: list):
    non_repeated = set(numbers)
    return len(non_repeated)

how_many_different_numbers([1, 2, 3, 1, 2, 3, 4, 1])

```
